```python
# 2ND LAB DA
import pandas as pd
health_info2=pd.read_csv("https://raw.githubusercontent.com/MarinaPandaCodes/Data-Science/refs/heads/main/info2.csv",header=0,sep=',')
print(health_info2)
```

```python
import matplotlib.pyplot as plt
health_info2.plot(x="Average_Pulse",y="Calorie_Burnage",kind='scatter')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()
```

```python
def slope(x1,y1,x2,y2):
    m=(y2-y1)/(x2-x1)
    return m
print(slope(80,240,90,260))
```

```python
import numpy as np
x=health_info2["Average_Pulse"]
y=health_info2["Calorie_Burnage"]
slop_intersept=np.polyfit(x,y,1)
print(slop_intersept)
```

```python
def Calorie_Burnage(x1,slop,intersept):
    s=(slop*x1)+intersept
    return s
print(Calorie_Burnage(125,2,80))
```

```python
import matplotlib.pyplot as plt
health_info2.plot(x="Average_Pulse",y="Calorie_Burnage",kind='line')
plt.ylim(ymin=0,ymax=400)
plt.xlim(xmin=0,xmax=160)
plt.show()
```
