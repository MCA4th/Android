```python
# 4TH LAB DA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
health_info3=pd.read_csv("https://raw.githubusercontent.com/MarinaPandaCodes/Data-Science/refs/heads/main/info3.csv",header=0,sep=',')
print(health_info3)
```

```python
from scipy import stats
x = health_info3["Average_Pulse"]
y = health_info3["Calorie_Burnage"]
slope,intercept,r,p,std_err=stats.linregress(x,y)
print("slope",slope)
print("intercept",intercept)
print("rvalue",r)
print("pvalue",p)
print("standarderr",std_err)
```

```python
def myfunc(x):
  return slope * x + intercept
new_y=list(map(myfunc,x))
print(new_y)

```

```python
plt.scatter(x,y)
plt.plot(x,new_y)
plt.ylim(ymin=0,ymax=2000)
plt.xlim(xmin=0,xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.scatter(x,y)
plt.plot(x,myfunc(x))
plt.show()
```

```python
import statsmodels.formula.api as smf
model=smf.ols("Calorie_Burnage ~ Average_Pulse",data=health_info3)
result=model.fit()
print(result.summary())
```
